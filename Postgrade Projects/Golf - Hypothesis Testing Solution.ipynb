```python
def sample_power_difftest(d,s,power=0.8,sig=0.05):
    z = scipy.stats.norm.isf([sig/2])
    zp = -1 * scipy.stats.norm.isf([power])
    n = (2*(s**2)) * ((zp+z)**2) / (d**2)
    return int(round(n[0]))
```
